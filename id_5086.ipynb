```python
# first submit

input_line = []
input_list = list(map(int,input().split()))

for i in range(int(len(input_list)/2)-1):
  num1 = input_list[2*i]
  num2 = input_list[1+2*i]

  if num1+num2 == 0:
    break
  else:
    if num1 > num2 :
      div = num1%num2
      if div == 0:
        print("multiple")
      else:
        print("neither")

    elif num2 > num1:
      div = num2%num1
      if div == 0:
        print("factor")
      else:
        print("neither")
    else:
      print("neither") 
```

```python
# correct code
while True:
    a, b = map(int, input().split())
    if a+b == 0:
        break
    if b % a == 0:
        print('factor')
    elif a % b == 0:
        print('multiple')
    else:
        print('neither')
```

```python
# final submit
while True:
  num1, num2 = map(int, input().split())
  if num1+num2 == 0:
    break
  else:
    if num1 > num2 :
      if num1%num2 == 0:
        print("multiple")
      else:
        print("neither")

    elif num2 > num1:
      if num2%num1 == 0:
        print("factor")
      else:
        print("neither")
    else:
      print("neither") 
```
